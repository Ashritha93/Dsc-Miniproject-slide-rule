```python
#Import json data amd storing it as a dataframe
df = pd.read_json('C:/Users/ASHRITHA/Downloads/data_wrangling_json (4)/data_wrangling_json/data/world_bank_projects.json')
df.head()
```

```python
#cleaning the data. In this case specific column () for the excercise
# Work around code to get unique values and dropping value pairs with name ==''
prjt_series = df['mjtheme_namecode'].apply(pd.Series).stack().reset_index(drop=True)
mask = prjt_series.apply(lambda x: x['name']!='')
uniq_val = prjt_series[mask].value_counts()
```

```python
#Getting the values of the index to list and acquiring a dict where value of 'code'  os the key and value of 'name' is the value
uniq_val = pd.DataFrame(uniq_val)
temp_list = uniq_val.index.to_list()
val_dict = {temp_list[i]['code'] : temp_list[i]['name'] for i in range(len(temp_list))}
val_dict
```

```python
# Replacing missing 'name' values in the column using the val_dict as reference.
def replace_vals(col):
    #print(col)
    for element in col:
        for i in range(len(element)):
            if(element[i]['name']==''):
                element[i]['name'] = val_dict[element[i]['code']]
replace_vals(df.mjtheme_namecode)
```
